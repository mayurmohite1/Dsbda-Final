```python
# Select only the columns of interest
selected_columns = ['1st', '2nd', '3rd', '4th', '5th']
df_selected = df[selected_columns].copy()
```

```python
Q1 = df_selected.quantile(0.25)
Q3 = df_selected.quantile(0.75)

# Calculate the interquartile range (IQR)
IQR = Q3 - Q1

# Define the upper and lower bounds for outlier detection
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = (df_selected < lower_bound) | (df_selected > upper_bound)
```

```python
outliers
```

```python
# Count the number of outliers in each column
outliers_count = outliers.sum()
# Print the count of outliers
print("Number of outliers in each column:")
print(outliers_count)
```

```python
# Identify which rows contain outliers in the selected columns
outlier_rows = df_selected[outliers.any(axis=1)]
# Print the rows containing outliers
print("Rows containing outliers in the selected columns:")
print(outlier_rows)
```
